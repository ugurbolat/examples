```python
circuit = qiskit.QuantumCircuit(4,4)
thetat = qiskit.circuit.Parameter('Temp')
thetax = qiskit.circuit.Parameter('X')
thetay = qiskit.circuit.Parameter('Y')
thetaz = qiskit.circuit.Parameter('Z')

circuit.ry(thetat,0)
circuit.ry(thetax,1)
circuit.ry(thetay,2)
circuit.ry(thetaz,3)

theta = qiskit.circuit.Parameter('Î¸')

for i in range(2):
    for j in range(4):
        circuit.rx(theta,j)
    for j in range(4):
        if j < 3:
            circuit.cnot(j,j+1)
        else:
            circuit.cnot(j,0)
    circuit.barrier(range(4))

circuit.measure(range(4),range(4))
circuit.draw(output="mpl", plot_barriers=False)
        
```
